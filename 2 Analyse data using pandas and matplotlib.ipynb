This is a dataframe giving information on the planets in the Solar system, downloaded using the seaborn package. The 
mean, maximum and minimum values of the orbital periods, masses and distances of the planets have been found using the
describe function. The dataframe has been sorted according to orbital period in ascending order. The number of dataframes in each column have been printed. rows of the file state_abbreviations and state_areas have been merged. The notebook also has a plot of the orbital periods, masses and distances of the planets at the end.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('planets')
```

```python
df
```

```python
df.describe()
```

```python
df.sort_values(by = 'orbital_period')
```
